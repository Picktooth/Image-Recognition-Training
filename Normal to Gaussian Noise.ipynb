# Add Gaussian Noise
#### The following code is used to add gaussian noise to a set of images 

```python
# Basics
import torch
import torchvision
import torchvision.transforms as transforms
import numpy as np
import pandas as pd
import os
import cv2

# Adding noise
from skimage.util import random_noise
from torchvision.utils import save_image

print("Success")
```

```python
BATCH_SIZE = 1
NUM_CLASSES = 10
GENERATIONS = 10000
TEST = GENERATIONS//10

transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

```python
"""
Assigning the training image set to be modified; A set of test images can be created from modified training images
"""

# Data Downloading
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
# Data Loader
trainLoader = torch.utils.data.DataLoader(trainset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)

# Image Categories
classes = ('airplane', 'automobile', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck') 
```

```python

# FUNCTIONS

# Apply gaussian noise to images in the training set
def gaussian_noise_train():
    for data in trainLoader:
        img, labels = data[0], data[1]
        gauss_img = torch.tensor(random_noise(img, mode='gaussian', mean=0, var=0.02, clip=True)) # Change var to affect noisyness
        

        
        # Grabbing image labels (Truth)
        print(' '.join(f'{classes[labels[j]]:5s}' for j in range(BATCH_SIZE)))
        label = ''.join(f'{classes[labels[j]]:5s}' for j in range(BATCH_SIZE))
        
        # Check amount of files in label class folder; If value < Target, add photo to folder
        if len([entry for entry in os.listdir(fr"G:\Program Files (x86)\Documents\Projects\Image Recognition Training\CIFAR10 Work\Noiseimg\train\{label.strip()}")
        if os.path.isfile(os.path.join(fr"G:\Program Files (x86)\Documents\Projects\Image Recognition Training\CIFAR10 Work\Noiseimg\train\{label.strip()}", entry))]) != TEST:
            torchvision.utils.save_image(gauss_img, fr"G:\Program Files (x86)\Documents\Projects\Image Recognition Training\CIFAR10 Work\Noiseimg\train\{label.strip()}\{i}_gaussian.png")
          # torchvision.utils.save_image(img, fr"C:\Users\John\Documents\Visual Studio Code Python\Image_Classification_Training\CIFAR10 Gaussian Noise\Noiseimg\{i}_img.png")

        else:
            print('Folder for {label} is full, discarding image')
        break

# Create a test set of noisy images by applying gaussian noise to images in the training set
def gaussian_noise_test():
    for data in trainLoader:
        img, labels = data[0], data[1]
        gauss_img = torch.tensor(random_noise(img, mode='gaussian', mean=0, var=0.007, clip=True))
        

        
        # Grabbing image labels (Truth)
        print(' '.join(f'{classes[labels[j]]:5s}' for j in range(BATCH_SIZE)))
        label = ''.join(f'{classes[labels[j]]:5s}' for j in range(BATCH_SIZE))
        
        # Check amount of files in label class folder; If value < Target, add photo to folder
        if len([entry for entry in os.listdir(fr"G:\Program Files (x86)\Documents\Projects\Image Recognition Training\CIFAR10 Work\Noiseimg\test\{label.strip()}")
        if os.path.isfile(os.path.join(fr"G:\Program Files (x86)\Documents\Projects\Image Recognition Training\CIFAR10 Work\Noiseimg\test\{label.strip()}", entry))]) != TEST:
            torchvision.utils.save_image(gauss_img, fr"G:\Program Files (x86)\Documents\Projects\Image Recognition Training\CIFAR10 Work\Noiseimg\test\{label.strip()}\{i}_gaussian.png")
        else:
            print('Folder for {label} is full, discarding image')
        break
    
# Check the number of images in train or test folder
def directory_count():
    count = 0
    
    # Change depending on whether you're generating images for train or test
    dir_path = './Noiseimg/train'
    for root_dir, cur_dir, files in os.walk(dir_path):
        count += len(files)
    return count
```

```python
# For generating train data
i = 0
while i < GENERATIONS:
    print(i)
    gaussian_noise_train()
    i = directory_count()
```

```python
# For generating test data
i = 0
while i < GENERATIONS:
    print(i)
    gaussian_noise_test()
    i = directory_count()
```
